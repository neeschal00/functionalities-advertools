```python
import pandas as pd
from advertools import (
    url_to_df,
    stopwords
)
import re
from collections import Counter
```

```python
from urllib.parse import urlparse, parse_qs
```

```python
crawl_df = pd.read_json("seo_crawler.jl",lines=True)
```

```python
url_df = url_to_df(crawl_df["url"])
url_df
```

```python
parse_qs(urlparse(url_df["url"][0]).query)
```

```python
url_df['params'] = url_df['url'].apply(lambda x: parse_qs(urlparse(x).query))
```

```python
url_df["params"]
```

```python
# URL word patterns
url_df['path'] = url_df['path'].fillna(" ")
url_df['url_words'] = url_df['path'].apply(lambda x: x.split('/')[1:])
url_df['url_words']
```

```python
# url_df['url_words'] = url_df['url_words'].fillna(" ")
url_df['word_counts'] = url_df['url_words'].apply(len)
url_df["word_counts"]
```
